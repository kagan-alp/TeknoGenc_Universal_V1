```python
# Tekno Genç Fizik Motoru V1.0
print("--- Tekno Genç Kuvvet Hesaplayıcı ---")

# Veri Girişi (Girdi)
kutle = float(input("Makinenin kütlesini gir (kg): "))
ivme = float(input("Makinenin ivmesini gir (m/s2): "))

# Fiziksel İşlem (Newton'un 2. Kanunu: F = m * a)
kuvvet = kutle * ivme

# Çıktı
print("-------------------------------------")
print(f"Hesaplanan Gereken Kuvvet: {kuvvet} Newton")
print("Durum: Fiziksel motor kararlı.")
```

```python
# Tekno Genç: Öğrenen Motor Deneyi
import numpy as np

# Veriler (Geçmiş tecrübelerimiz gibi düşün)
# Mesela: 1kg kütle ve 2m/s2 ivme olunca sonuç 2N olmuş.
veriler = np.array([2, 4, 6, 8, 10], dtype=float) # Sonuçlar (Kuvvetler)
girdiler = np.array([1, 2, 3, 4, 5], dtype=float) # Girdiler (Kütle x İvme çarpımı)

# Yapay zekaya diyoruz ki: "Bu sayılar arasında bir ilişki var, bul!"
# Biz formülü söylemiyoruz, o bakarak çözecek.
tahmin = 7 # Mesela kütle*ivme = 7 ise kuvvet ne olur?
sonuc = np.interp(tahmin, girdiler, veriler)

print(f"Yapay Zeka Tahmini: {tahmin} birimlik girdi için sonuç {sonuc} Newton olmalı.")
print("Tekno Genç AI: Formülü ben buldum!")
```

```python
# Tekno Genç: Veri Görselleştirme Motoru
import matplotlib.pyplot as plt

# Makine Verileri (Örnek: Zaman içindeki sıcaklık veya basınç)
zaman = [1, 2, 3, 4, 5, 6, 7]
verimlilik = [10, 25, 45, 30, 60, 85, 95] # Makinenin performans artışı

plt.plot(zaman, verimlilik, marker='o', color='red', linestyle='--')
plt.title("Tekno Genç: Makine Performans Analizi")
plt.xlabel("Zaman (Gün)")
plt.ylabel("Verimlilik (%)")
plt.grid(True)

print("Grafik oluşturuluyor... Tekno Genç raporu hazır!")
plt.show()
```
